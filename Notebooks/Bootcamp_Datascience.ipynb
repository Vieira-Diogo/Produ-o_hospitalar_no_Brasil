```python
import pandas as pd

URL = 'https://raw.githubusercontent.com/Vieira-Diogo/Producao_hospitalar_no_Brasil/main/Dados/A001558189_28_143_208.csv'
```

```python
dados = pd.read_csv(URL, encoding="ISO-8859-1", skiprows=3, sep=";", skipfooter=12, engine= 'python', thousands= ".", decimal=",")
dados.head()
```

```python
dados.tail()
    
```

```python
dados.mean()
```

```python
pd.options.display.float_format = "{:.2f}".format
```

```python
dados.mean()
```

```python
dados["2008/Ago"].mean()
```

```python
dados.plot(x="Unidade da Federação", y="2008/Ago", kind = "bar", figsize=(9,6)
)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind = "bar", figsize=(9,6))
ax.yaxis.set_major_formatter (ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Gastos com procedimentos hospitalares do SUS")
plt.show()
```

```python
ax = dados.plot(x="Unidade da Federação", y="2021/Mar", kind = "bar", figsize=(9,6))
ax.yaxis.set_major_formatter (ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Gastos com procedimentos hospitalares do SUS")
plt.show()
```
